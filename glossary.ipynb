```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


site = requests.get ('https://centerforbookarts.org/resources/glossary')
html_source = site.content
soup = BeautifulSoup(html_source)
soup


```

```python
object_type = []
definition = []

for h4_tag in soup.find_all('h4', class_="wp-block-heading"):
    # Get the text from the <h4> tag for object_type
    object_type_text = h4_tag.get_text(strip=True)
    object_type.append(object_type_text)
    
    # Find the next sibling <p> tag and get its text for definition
    p_tag = h4_tag.find_next_sibling('p')
    if p_tag:
        definition_text = p_tag.get_text(strip=True)
        definition.append(definition_text)
    else:
        # If there's no <p> following the <h4>, append None or empty string
        definition.append("")

# Display the results
for obj, defn in zip(object_type, definition):
    print(f"Object Type: {obj}")
    print(f"Definition: {defn}\n")
```

```python
import csv

df = pd.DataFrame({'object_type': object_type, 'definition': definition})
df.to_csv('object_types.csv', index=False, header=True)
```
